### Question 1.3
Split data in training and test set. Do not use the test-set unless for a final evaluation in 1.6.
For each categorical variable, cross-validate a Linear Regression model using just this variable
(one-hot-encoded). Visualize the relationship of the categorical variables that provide the best
R^2 value with the target.

```python
cat_attrbs = list(df.select_dtypes(include=['object']).columns)

num_attrbs = list(df.select_dtypes(exclude=['object']).columns)
num_attrbs = num_attrbs[2:]
num_attrbs.remove('SalePrice')


for attrb in num_attrbs:
    df[attrb] = df[attrb].astype(float)

# Fill NaN in numeric columns with 0s
values = {column:0 for column in num_attrbs}
df.fillna(value=values,inplace=True)

values = {column:'Empty' for column in cat_attrbs}
df.fillna(value=values,inplace=True)
```

```python
df.head()
```

```python
scores = {}

y = df['SalePrice']
x = df.drop(columns=['SalePrice'])

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)

for attrb in cat_attrbs:
    # Build One Hot encoder for the categorial variable.
    enc = OneHotEncoder(categories='auto')
    feature = np.array(x_train[attrb])
    encoded = enc.fit_transform(feature.reshape(-1,1))

    # Build a Linear Regression model.
    reg = LinearRegression()
    #cv = cross_val_score(reg, encoded, y, cv=5)
    cv_results = cross_validate(reg, encoded.toarray(), y_train, scoring='r2', cv=3)

    # Calculate R^2 
    scores[attrb] = cv_results['test_score'].mean()

best = max(scores,key=scores.get)
print("The categorial property feature with the highest R^2 value when cross-validated with the price: {} with {}".format(best, max(scores.values())))
print(scores)
```

```python
enc = OneHotEncoder(categories='auto')
feature = np.array(x[best])
encoded = enc.fit_transform(feature.reshape(-1,1)).toarray()


encoded = [''.join(list(map(lambda x: str(x),list(e.astype(int))))) for e in encoded]
encoded = [int(e, 2) for e in encoded]

coded = df.copy()
coded[best] = encoded
scatter_matrix(coded[[best,'SalePrice']])
```

### Question 1.4
Use ColumnTransformer and pipeline to encode categorical variables. Evaluate Linear
Regression (OLS), Ridge, Lasso and ElasticNet using cross-validation with the default
parameters. Does scaling the data (within the pipeline) with StandardScaler help?

```python
# Without scaling

# Preprocessing categorial features
cat_transformer = OneHotEncoder(handle_unknown='ignore')
preprocessor = ColumnTransformer(transformers=[('cat', cat_transformer, cat_attrbs)])

models = [('Linear Regression',LinearRegression()), ('Ridge Regression',Ridge()), ('Lasso',Lasso()), ('Elastic Net',ElasticNet())]
scores = {}


for model in models:
    # Appending the model to the preprocessing pipeline.
    clf = Pipeline(steps=[('preprocessor',preprocessor),
                      ('classifier', model[1])])
    scores[model[0]] = cross_val_score(clf, x_train, y_train, cv = 3).mean()

print(scores)
```

```python
# With scaling
numeric_transformer = Pipeline(steps=[('scalar',StandardScaler())])
preprocessor = ColumnTransformer(transformers=[('num', numeric_transformer , num_attrbs),('cat', cat_transformer, cat_attrbs)])


scores1 = {}
for model in models: 
    clf = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', model[1])])
    scores1[model[0]] = cross_val_score(clf, x_train, y_train, cv=3).mean()

print(scores1)
```

We observe there is an increase in the score of the models after adding scaling with Elastic Net having the biggest impact of scaling and afterwards Lasso.

### Question 2.2

Split data into training and test set. Build a pipeline for dealing with categorical variables.
Evaluate Logistic Regression and nearest centroids using
cross-validation. How different are the results? How does scaling the continuous features with
StandardScaler influence the results?

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import NearestCentroid

cat_attrbs = list(data.select_dtypes(include=['object']).columns)
cat_attrbs.remove('Churn')

num_attrbs = list(data.select_dtypes(include=['number']).columns)

for attrb in num_attrbs:
    data[attrb] = data[attrb].astype(float)

x = data.drop(columns=['Churn'])
y = data['Churn']

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
# Without scaling
preprocessor = ColumnTransformer(transformers=[('num', StandardScaler(), num_attrbs),('cat', cat_transformer, cat_attrbs)])

models = [('Logistic Regression',LogisticRegression()),('Nearest Centroid',NearestCentroid())]
scores = {}

parameters = {'logistic':{'logistic__C':[0.4,0.5,0.7,1.2,1.5]},
              'nearestcentroid':{'nearestcentroid__shrink_threshold':[0.6,None]}}

for model in models:
    # Appending the model to the preprocessing pipeline.
    clf = Pipeline(steps=[('preprocessor',preprocessor),
                      ('classifier', model[1])])
    scores[model[0]] = cross_val_score(clf, x_train, y_train, cv = 3).mean()

print(scores)
```

```python
# With scaling
scores1 = {}
for model in models: 
    clf = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', model[1])])
    scores1[model[0]] = cross_val_score(clf, x_train, y_train, cv=3).mean()

print(scores1)
```

### Question 2.3

Tune the parameters using GridSearchCV. Do the results improve?

```python
clf = Pipeline(steps=[('preprocessor',preprocessor),
                      ('logistic', LogisticRegression())])
clf = GridSearchCV(clf, parameters['logistic'], cv=3)
clf.fit(x_train,y_train)
```

```python
print("Best score: {} with parameter: {}".format(clf.best_score_, clf.best_params_))
print(clf.cv_results_['mean_test_score'])
plt.plot([0.4,0.5,0.7,1.2,1.5],clf.cv_results_['mean_test_score'])
```

```python
categorical = x_train.dtypes == object
models = ['LogisticRegression', 'NearestCentroid']

scale_trans = make_column_transformer( (StandardScaler(), ~categorical), (OneHotEncoder(handle_unknown='ignore'), categorical) )
parameters = {'LogisticRegression':{'model__C' : np.logspace(-4, 4, 20)}, 
              'NearestCentroid': {'model__shrink_threshold': [0.01, 0.1, 0.15, 0.2, 0.3, 0.4, None]}}
result = {}
for model in models:
    pipe2 = Pipeline(steps=[('preprocess', scale_trans), ('model', eval(model + '()'))])
    result[model] = GridSearchCV(pipe2, param_grid=parameters[model], cv=4, return_train_score=True)
    result[model].fit(x_train, y_train)
```

```python
clf = Pipeline(steps=[('preprocessor',preprocessor),
                      ('nearestcentroid', NearestCentroid())])
clf = GridSearchCV(clf, parameters['nearestcentroid'], cv=3)
clf.fit(x_train,y_train)
print(clf.best_score_, clf.best_params_)
print(clf.cv_results_['mean_test_score'])
#plt.plot([0.4,0.5,0.7,1.2,1.5],clf.cv_results_['mean_test_score'])
```

### Question 2.4

Change the cross-validation strategy from ‘stratified k-fold’ to ‘kfold’ with shuffling. Do the
parameters that are found change? Do they change if you change the random seed of the
shuffling? Or if you change the random state of the split into training and test data?

```python
from sklearn.model_selection import KFold

# We used shuf fled KFold
kf = KFold(shuffle=True)

clf = Pipeline(steps=[('preprocessor',preprocessor),
                      ('logistic', LogisticRegression())])
clf = GridSearchCV(clf, parameters['logistic'], cv=kf)
clf.fit(x_train,y_train)
```

```python
print("Best score: {} with parameter: {}".format(clf.best_score_, clf.best_params_))
print(clf.cv_results_['mean_test_score'])
plt.plot([0.4,0.5,0.7,1.2,1.5],clf.cv_results_['mean_test_score'])
```

We observe the best parameter is changed alongside with the score which increased by factor of 2*10^(-3) upon cross-validating with shuffle.
